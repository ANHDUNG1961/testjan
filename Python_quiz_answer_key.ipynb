```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/players (1).csv")
```

```python
df
```

```python
plt.scatter(df['height'], df['weight'])






# Customize the plot
plt.xlabel('height')
plt.ylabel('weight')
plt.title('Scatter Plot height vs. weight')


# Show the plot
plt.show()
```

```python
type(df['height'][0])
```

```python
df['feet'] = df['height'].apply(lambda x: int(x[0]))
```

```python
df['inches'] = df['height'].apply(lambda x: int(x[2:]))
```

```python
df.head(5)
```

```python
df["height in inches"] = df['feet']*12 + df['inches']
```

```python
plt.scatter(df['height in inches'], df['weight'])






# Customize the plot
plt.xlabel('height')
plt.ylabel('weight')
plt.title('Scatter Plot height vs. weight')


# Show the plot
plt.show()
```
